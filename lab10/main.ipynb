```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return np.sqrt(x)

def L2(x):
    l0 = ((x-4)*(x-9)) / ((1-4)*(1-9))
    l1 = ((x-1)*(x-9)) / ((4-1)*(4-9))
    l2 = ((x-1)*(x-4)) / ((9-1)*(9-4)) 
    return 1*l0 + 2*l1 + 3*l2
   
def N(x):
    return 1 + (1/3)*(x-1) - (1/60)*(x-1)*(x-4)   
 
plt.figure(figsize=(10, 6))
x = np.linspace(1, 9, 100)
plt.plot(x, f(x), 'r-', label='f(x) = sqrt(x)')
plt.plot([1, 4, 9], [1, 2, 3], 'ro', markersize=8, label='Data points')
plt.plot(x, L2(x), 'b-', label='L2(x)')
plt.plot(x, N(x), 'g--', label='N(x)')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
print(f(x))
print(L2(x))
print(L2(2))
print(N(2))
```
