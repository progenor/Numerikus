# Newton

```python
import matplotlib.pyplot as plt
import numpy as np

x = [-2, -1, 0 , 1, 2]
y = [-15, -4, -1, 0, 5]

def newton_coeff(x, y):
    n = len(x)
    a = y.copy()
    for j in range(1, n):
        for i in range(n-1, j-1, -1):
            a[i] = (a[i] - a[i-1]) / (x[i] - x[i-j])
    return a

def newton_eval(x, a, value):
    n = len(a) - 1
    result = a[n]
    for i in range(n-1, -1, -1):
        result = result * (value - x[i]) + a[i]
    return result

a = newton_coeff(x, y)
```

```python

x_dense = np.linspace(min(x), max(x), 100)
y_dense = [newton_eval(x, a, xi) for xi in x_dense]

plt.plot(x, y, 'ro', label='x,y')
plt.plot(x_dense, y_dense, 'b-', label='Newton')
plt.legend()
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# Lagrange

```python
# Egyszerű Lagrange interpolációs polinom Pythonban

# Adatok (x és y értékek)
x = [-2, -1, 0 , 1, 2]
y = [-15, -4, -1, 0, 5]

# Lagrange interpolációs polinom kiértékelése
def lagrange_interp(x, y, value):
    total = 0
    n = len(x)
    for i in range(n):
        term = y[i]
        for j in range(n):
            if i != j:
                term *= (value - x[j]) / (x[i] - x[j])
        total += term
    return total
```

```python
plt.plot(x, y, 'ro', label='x,y')
x_dense = np.linspace(min(x), max(x), 100)
y_dense = [lagrange_interp(x, y, xi) for xi in x_dense]
plt.plot(x_dense, y_dense, 'g-', label='Lagrange')
plt.legend()
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
