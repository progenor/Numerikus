```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 3*x*(1-x)

ax = plt.subplot(111)

x0 = 0.2
i=0
while i < 1000:
    x1 = f(x0)
    ax.plot([x0, x0], [x0, x1], 'r-')
    ax.plot([x0, x1], [x1, x1], 'r-')
    x0 = x1
    
    
    i+=1

x_vals = np.linspace(0, 1, 100)
y_vals = f(x_vals)
ax.plot(x_vals, y_vals, 'b-', label='f(x) = 3x(1-x)')
ax.plot([0,1], [0,1], 'g-', label='y=x')
plt.show()

```
