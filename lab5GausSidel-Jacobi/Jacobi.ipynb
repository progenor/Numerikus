# Jacobi iteracio

$$x^{k+1}=D^{-1}(L+U)x^{(k)}+D^{-1}b)$$

```python
import numpy as np
import matplotlib.pyplot as plt


A = np.array([[6, 2, 3], [3, 7, 3], [5, 6, 12]])
b = np.array([[1], [2], [3]])
N = 1000
Eps = 1e-6
x0 = np.zeros((3, 1))
```

```python

# Jacobi iteration
y = x0.copy()  # Save current solution in y
x = np.zeros((3, 1))  # New solution
errors = []  # To store convergence history

for k in range(N):
    for i in range(3):
        s = 0
        for l in range(3):
            if l != i:
                s += A[i, l] * y[l, 0]
        x[i, 0] = (b[i, 0] - s) / A[i, i]
    
    # Check convergence
    error = np.linalg.norm(x - y)
    errors.append(error)
    
    if error < Eps:
        print(f"Converged after {k+1} iterations")
        break
    
    # Update y for the next iteration
    y = x.copy()
else:
    print(f"Did not converge after {N} iterations")

# Print result
print("Solution:", x.flatten())

# Plot convergence history
plt.figure(figsize=(10, 6))
plt.semilogy(range(1, len(errors) + 1), errors)
plt.xlabel('Iteration')
plt.ylabel('Error (log scale)')
plt.title('Jacobi Method Convergence')
plt.grid(True)
plt.show()
```

```python

# Solve directly with numpy
solution = np.linalg.solve(A, b)
print(f"Solution: {solution.flatten()}")
```
