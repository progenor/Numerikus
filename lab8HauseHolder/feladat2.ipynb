```python
import math

def f(x):
    return x**3 - 10*x**2 + 5

def f_deriv(x):
    return 3*x**2 - 20*x

def f_2_deriv(x):
    return 6*x - 20

def householder_method(x0, d=3, tol=1e-10, max_iter=20):
    x = x0
    for i in range(max_iter):
        fx = f(x)
        fpx = f_deriv(x)
        
        if abs(fx) < tol:
            print("megvan megoldas")
            return x
            
        if fpx == 0:
            print("derivative is zero")
            break
            
        correction = fx / fpx
        
        if d == 1:
            x_new = x - correction
        elif d == 2:
            fppx = f_2_deriv(x)
            factor = 1 / (1 - (fx * fppx) / (2 * fpx**2))
            x_new = x - correction * factor
        elif d == 3:
            fppx = f_2_deriv(x)
            factor = 1 + (fx * fppx) / (2 * fpx**2)
            x_new = x - correction * factor
        else:
            raise ValueError("d must be 1, 2, or 3")

        print(f"iter {i+1}: x = {x_new:.10f}, f(x) = {f(x_new):.3e}")

        if abs(x_new - x) < tol:
            print("konvergal!")
            return x_new
        
        x = x_new

    print("nem erte el")
    return x

initial_guess = 0.7

print("d=1:")
solution1 = householder_method(initial_guess, d=1)
print(f"x = {solution1:.10f}")

print("\nd=2:")
solution2 = householder_method(initial_guess, d=2)
print(f"x = {solution2:.10f}")

print("\nd=3:")
solution3 = householder_method(initial_guess, d=3)
print(f"x = {solution3:.10f}")
```
