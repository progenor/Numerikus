```python
import numpy as np
n = 3
A = [[4, 2, 14], [2, 17, -5], [14, -5, 83]]
L = [[0, 0, 0], [0, 0, 0], [0, 0, 0]]
```

```python
for j in range(n):
    osszeg = 0
    for k in range(j-1):
        osszeg += L[j][k]**2
    L[j][j] = (A[j][j] - osszeg)**0.5
    for i in range(j+1, n):
        osszeg = 0
        for k in range(j-1):
            osszeg += L[i][k]*L[j][k]
        L[i][j] = (A[i][j] - osszeg)/L[j][j]
```

```python
l = np.zeros((len(A), n))
print(l)
```
